```python
# import dependencies
import pandas as pd
```

```python
# Define a function that reads a text file with two columns and stores it in a dictionary.
def read_file_to_dict(file_path):
    """
    Reads a text file with two columns and stores it in a dictionary.
    
    Args:
    file_path (str): The path to the text file.
    
    Returns:
    dict: A dictionary with keys from the first column and values from the second column.
    """
    data_dict = {}
    
    with open(file_path, 'r') as file:
        for line in file:
            # Strip leading/trailing whitespace and split by whitespace
            parts = line.strip().split(maxsplit=1)
            if len(parts) == 2:
                key, value = parts
                data_dict[key] = value
    
    return data_dict
```

```python
#Use the function to read the state  data into dictionaries.

state_dict = read_file_to_dict('resources/state_codes.txt')
print(state_dict)
```

```python
# for key in state_dict: if key begins with 0, then the '0' should not be used for mapping the state code to the state name.
# Create a new dictionary with the correct state codes.
state_dict_new = {}
for key, value in state_dict.items():
    if key[0] == '0':
        state_dict_new[key[1:]] = value
    else:
        state_dict_new[key] = value 
display(state_dict_new)
```

```python
# Read the birth data from csv file and store it in a dataframe
birth_data_df = pd.read_csv('resources/allBirthData.csv')
display(birth_data_df.head())
display(birth_data_df.tail())
birth_data_df.info()
```

```python
# convert the state and county columns to strings
birth_data_df['State'] = birth_data_df['State'].astype(str)
birth_data_df['County'] = birth_data_df['County'].astype(str)
birth_data_df.info()
```

```python
# create a new column 'StateName' by mapping the 'State' column to the state names using the state_dict
birth_data_df['StateName'] = birth_data_df['State'].map(state_dict_new)
display(birth_data_df.head())
display(birth_data_df.tail())
```

```python
#delete Unnamed: 0 column and the index column
birth_data_df.drop(columns=['Unnamed: 0'], inplace=True)
display(birth_data_df.head())
```

```python
# Drop columns County and countyBirths
birth_data_df.drop(columns=['County', 'countyBirths'], inplace=True)
```

```python
# select data from 1992 onwards only
birth_data_df = birth_data_df[birth_data_df['Year'] >= 1992]
```

```python
# select data for the states of New York, New Jersey, and Connecticut only
ny_nj_ct_birth_data_df = birth_data_df[birth_data_df['StateName'].isin(['NEW YORK', 'NEW JERSEY', 'CONNECTICUT'])]
display(ny_nj_ct_birth_data_df.head())
ny_nj_ct_birth_data_df.info()
```

```python
# select data for all states except New York, New Jersey, and Connecticut
other_birth_data_df = birth_data_df[~birth_data_df['StateName'].isin(['NEW YORK', 'NEW JERSEY', 'CONNECTICUT'])]
display(other_birth_data_df.head())
other_birth_data_df.info()
```

```python
# drop the index column
ny_nj_ct_birth_data_df.drop(columns=['index'], inplace=True)
other_birth_data_df.drop(columns=['index'], inplace=True)
# reset the index for the dataframes
ny_nj_ct_birth_data_df.reset_index(drop=True, inplace=True)
other_birth_data_df.reset_index(drop=True, inplace=True)


display(ny_nj_ct_birth_data_df.head())
display(other_birth_data_df.head())
```

```python
# change the StateName to title case
ny_nj_ct_birth_data_df['StateName'] = ny_nj_ct_birth_data_df['StateName'].str.title()
other_birth_data_df['StateName'] = other_birth_data_df['StateName'].str.title()
ny_nj_ct_birth_data_df.head()
other_birth_data_df.head()
```

```python
# change the Month column using the Year and Month columns to a datetime object with the day as the first day of the month

ny_nj_ct_birth_data_df['Month'] = pd.to_datetime(ny_nj_ct_birth_data_df['Year'].astype(str) + '-' + ny_nj_ct_birth_data_df['Month'].astype(str) + '-01')
other_birth_data_df['Month'] = pd.to_datetime(other_birth_data_df['Year'].astype(str) + '-' + other_birth_data_df['Month'].astype(str) + '-01')
ny_nj_ct_birth_data_df.head()
other_birth_data_df.head()
```

```python
# drop the StateName and State column from the ny_nj_ct_birth_data_df dataframe
ny_nj_ct_birth_data_df.drop(columns=['StateName' , 'State'], inplace=True)
ny_nj_ct_birth_data_df.head()
```

```python
# show data for ny nj ct for only Jan 1992
temp_df = ny_nj_ct_birth_data_df[(ny_nj_ct_birth_data_df['Month'] == '1992-01-01')]

# what is the sum of the births for the states of New York, New Jersey, and Connecticut in January 1992?
temp_df['stateBirths'].sum()
```

```python
# create a new dataframe with the columns Year, Month, Births, where Births is the sum of the monthly births for ny_nj_ct_birth_data_df
tri_state_births_df = ny_nj_ct_birth_data_df.groupby(['Year', 'Month'])['stateBirths'].sum().reset_index()
#rename the column stateBirths to TriStateBirths
tri_state_births_df.rename(columns={'stateBirths': 'TriStateBirths'}, inplace=True)

display(tri_state_births_df.head())
tri_state_births_df.info()
```

```python
# create a new dataframe with the columns Year, Month, Births, where Births is the sum of the monthly births for other_birth_data_df
other_monthly_births_df = other_birth_data_df.groupby(['Year', 'Month'])['stateBirths'].sum().reset_index()

# rename the column stateBirths to OtherBirths
other_monthly_births_df.rename(columns={'stateBirths': 'OtherBirths'}, inplace=True)    
display(other_monthly_births_df.head())
other_monthly_births_df.info()
```
