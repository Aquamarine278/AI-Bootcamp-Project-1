```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import employment.csv and read_csv in as a df
# import forecast_911_employment.csv and read_csv in as a df
# import forecast_katrina_employment and read_csv in as a df
# use display() to preview


employment_df = pd.read_csv('resources/employment.csv', index_col= 'ds')
forecast_911_employment_df = pd.read_csv('resources/forecast_911_employment.csv', index_col= 'ds')
forecast_katrina_employment_df = pd.read_csv('resources/forecast_katrina_employment.csv', index_col= 'ds')
display(employment_df.head())
display(forecast_911_employment_df.tail())
display(forecast_katrina_employment_df.tail())
```

```python
# import unemployment_cleaned.csv and read_csv in as a df
# import forecast_911_unemployment.csv and read_csv in as a df
# import forecast_katrina_unemployment and read_csv in as a df
# use display() to preview

unemployment_df = pd.read_csv('resources/unemployment_cleaned.csv', index_col= 'ds')
forecast_911_unemployment_df = pd.read_csv('resources/forecast_911_unemployment.csv', index_col= 'ds')
forecast_katrina_unemployment_df = pd.read_csv('resources/forecast_katrina_unemployment.csv', index_col= 'ds')

display(unemployment_df.head())
display(forecast_911_unemployment_df.tail())
display(forecast_katrina_unemployment_df.tail())
```

```python
# import S&P_cleaned.csv and read_csv in as a df
# import forecast_9_11_S&P.csv and read_csv in as a df
# import forecast_katrina_S&P and read_csv in as a df
# use display() to preview

sp500_df = pd.read_csv('resources/S&P_cleaned.csv', index_col= 'ds')
forecast_9_11_sp500_df = pd.read_csv('resources/forecast_9_11_S&P.csv', index_col= 'ds')
forecast_katrina_sp500_df = pd.read_csv('resources/forecast_katrina_S&P.csv', index_col= 'ds')
display(forecast_katrina_sp500_df.tail())
display(forecast_9_11_sp500_df.tail())
display(sp500_df.head())
```

```python
# import gdp_cleaned.csv and read_csv in as a df
# import forecast_9_11_gdp.csv and read_csv in as a df
# import forecast_katrina_gdp.csv and read_csv in as a df
# use display() to preview

gdp_df = pd.read_csv('resources/gdp_cleaned.csv', index_col= 'Date')
forecast_9_11_gdp_df = pd.read_csv('resources/forecast_9_11_gdp.csv', index_col= 'ds')
forecast_katrina_gdp_df = pd.read_csv('resources/forecast_katrina_gdp.csv', index_col= 'ds')

display(gdp_df.head())
display(forecast_9_11_gdp_df.tail())
display(forecast_katrina_gdp_df.tail())
```

```python
# slice sp500_df, and employment_df from 1988 to 2004
# slice gdp_df from 1992 to 2004
nine_eleven_sp500_df = sp500_df.loc['1988': '2004']
nine_eleven_employment_df = employment_df.loc['1988': '2004']
nine_eleven_gdp_df = gdp_df.loc['1992': '2004']
nine_eleven_unemployment_df = unemployment_df.loc['1988': '2004']
# plot them in the same graph
plt.figure(figsize=(15,7.5))
plt.plot(nine_eleven_sp500_df.index, nine_eleven_sp500_df['y'], label= 'S&P500')
plt.plot(nine_eleven_employment_df.index, nine_eleven_employment_df['y'], label = 'Employment')
plt.plot(nine_eleven_gdp_df.index, nine_eleven_gdp_df['Monthly Nominal GDP Index'], label = 'GDP')
plt.legend()
plt.show()
```

```python
# plot them by percentage change

nine_eleven_employment_df['percentage change'] = nine_eleven_employment_df['y'].pct_change()
nine_eleven_gdp_df['percentage change'] = nine_eleven_gdp_df['Monthly Nominal GDP Index'].pct_change()
nine_eleven_sp500_df['percentage change'] = nine_eleven_sp500_df['y'].pct_change()
nine_eleven_unemployment_df['percentage change'] = nine_eleven_unemployment_df['y'].pct_change()

plt.figure(figsize=(15,7.5))
plt.plot(nine_eleven_sp500_df.index, nine_eleven_sp500_df['percentage change'], label= 'S&P500')
plt.plot(nine_eleven_unemployment_df.index, nine_eleven_unemployment_df['percentage change'], label = 'unemployment')
plt.plot(nine_eleven_gdp_df.index, nine_eleven_gdp_df['percentage change'], label = 'GDP')
plt.legend()
plt.show()
```
